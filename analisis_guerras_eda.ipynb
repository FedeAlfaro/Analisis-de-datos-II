````python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Importar módulos personalizados
from data_loader import load_data
from data_explorer import DataExplorer
from data_cleaner import DataCleaner

# Configuración general
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

print("✓ Todas las librerías importadas correctamente")
print("\n📦 Librerías cargadas desde requirements.txt:")
print("  • pandas, numpy, matplotlib, seaborn, scipy, scikit-learn, openpyxl")
````

## 2. Cargar el Archivo CSV

````python
# Cargar el dataset usando la función load_data estandarizada
archivo = "Datos_Guerras.csv"
df = load_data(archivo, encoding='utf-8', verbose=True)

print(f"\nDataset cargado: {archivo}")
print(f"Dimensiones: {df.shape}")
print(f"\n{'='*80}")
print("Primeras 5 filas del dataset:")
print(f"{'='*80}")
print(df.head())
````

## 3. Exploración Básica de Datos

````python
print("Shape (dimensiones):", df.shape)
print("\n" + "="*80)
print("Últimas 5 filas del dataset:")
print("="*80)
print(df.tail())

print("\n" + "="*80)
print("Información general del dataset:")
print("="*80)
df.info()

print("\n" + "="*80)
print("Columnas disponibles:")
print("="*80)
for i, col in enumerate(df.columns, 1):
    print(f"  {i}. {col} ({df[col].dtype})")
````

## 4. Análisis de Tipos de Datos

````python
# Crear instancia de DataExplorer
explorer = DataExplorer(df, verbose=True)

# Obtener resumen estructural (tipos de datos, cardinalidad, memoria, etc.)
print("\n" + "="*80)
print("RESUMEN ESTRUCTURAL DEL DATASET")
print("="*80)
structural_summary = explorer.get_structural_summary(cardinality_threshold=50)
print(structural_summary.to_string())
````

## 5. Estadísticas Descriptivas

````python
print("\n" + "="*80)
print("ESTADÍSTICAS DESCRIPTIVAS")
print("="*80)
print(df.describe().T.to_string())

print("\n" + "="*80)
print("RECUENTO DE COLUMNAS POR TIPO")
print("="*80)
tipo_counts = df.dtypes.value_counts()
print(tipo_counts)

# Mostrar información adicional por tipo
print("\n" + "="*80)
print("COLUMNAS NUMÉRICAS:")
print("="*80)
numeric_cols = df.select_dtypes(include='number').columns.tolist()
print(f"Encontradas {len(numeric_cols)} columnas numéricas: {numeric_cols}")

print("\n" + "="*80)
print("COLUMNAS CATEGÓRICAS/TEXTO:")
print("="*80)
object_cols = df.select_dtypes(include='object').columns.tolist()
print(f"Encontradas {len(object_cols)} columnas de texto: {object_cols}")
````

## 6. Detección de Valores Faltantes y Duplicados

### 6.1 Análisis de Valores Faltantes

````python
# Análisis de valores nulos usando DataExplorer
nulls_df = explorer.analyze_nulls()
print("\n" + "="*80)
print("ANÁLISIS DE VALORES FALTANTES")
print("="*80)
print(nulls_df.to_string())
````

### 6.2 Análisis de Duplicados

````python
# Análisis de duplicados
duplicates_result = explorer.analyze_duplicates()

print("\n" + "="*80)
print("ANÁLISIS DE DUPLICADOS")
print("="*80)
print(f"Total de filas duplicadas (todas las columnas): {duplicates_result['total_exact_dups']}")
print(f"Porcentaje de duplicados: {duplicates_result['pct_exact_dups']:.2f}%")

if duplicates_result['partial_dups_df'] is not None and len(duplicates_result['partial_dups_df']) > 0:
    print("\nDuplicados parciales encontrados:")
    print(duplicates_result['partial_dups_df'].to_string())
else:
    print("\nNo se encontraron duplicados parciales")
````

## 7. Verificación de Formato Tidy

````python
# Verificar formato tidy
tidy_result = explorer.check_tidy_format()

print("\n" + "="*80)
print("VERIFICACIÓN DE FORMATO TIDY")
print("="*80)
print(f"¿Está en formato tidy? {tidy_result['is_tidy']}")

if not tidy_result['is_tidy']:
    print(f"\nProblemas encontrados ({len(tidy_result['issues'])}):")
    for i, issue in enumerate(tidy_result['issues'], 1):
        print(f"  {i}. {issue}")
    
    print(f"\nSugerencias ({len(tidy_result['suggestions'])}):")
    for i, sugg in enumerate(tidy_result['suggestions'], 1):
        print(f"  {i}. {sugg}")
else:
    print("✓ El dataset está correctamente estructurado en formato tidy")
````

## 8. Detección de Varianza Baja y Outliers

### 8.1 Columnas con Varianza Baja

````python
# Detectar columnas con varianza baja
low_variance_cols = explorer.detect_low_variance(threshold=0.01)

print("\n" + "="*80)
print("DETECCIÓN DE VARIANZA BAJA")
print("="*80)
if low_variance_cols:
    print(f"Se encontraron {len(low_variance_cols)} columnas con varianza baja:")
    for col in low_variance_cols:
        print(f"  - {col}")
else:
    print("✓ No se encontraron columnas con varianza baja")
````

### 8.2 Detección de Outliers

````python
# Detectar outliers
outliers_df = explorer.detect_outliers()

print("\n" + "="*80)
print("DETECCIÓN DE OUTLIERS (Método: IQR)")
print("="*80)

if len(outliers_df) > 0:
    print(f"Se encontraron {len(outliers_df)} valores atípicos")
    print("\nDetalles de outliers:")
    print(outliers_df.to_string())
else:
    print("✓ No se detectaron outliers en el dataset")
````

## 9. Análisis de Distribuciones y Normalidad

### 9.1 Gráfico de Normalidad (Q-Q y Histogramas)

````python
# Crear gráficos de normalidad
plt.figure(figsize=(16, 10))
explorer.plot_normality()
plt.tight_layout()
plt.show()

print("\n✓ Gráficos de normalidad creados")
````

## 10. Análisis de Correlaciones

### 10.1 Matriz de Correlación (Heatmap)

````python
# Crear matriz de correlación
fig = plt.figure(figsize=(14, 8))
explorer.plot_correlation_heatmap()
plt.tight_layout()
plt.show()

print("\n✓ Matriz de correlación creada")
````

````python
# Crear gráficos de dispersión por pares de variables numéricas
try:
    plt.figure(figsize=(16, 12))
    explorer.plot_scatter()
    plt.tight_layout()
    plt.show()
    print("\n✓ Gráficos de dispersión creados")
except Exception as e:
    print(f"Nota: No se pudieron generar gráficos de dispersión: {e}")
````

## 11. Resumen de Alertas y Problemas Detectados

````python
# Generar resumen de alertas
alerts_df = explorer.generate_alert_summary()

print("\n" + "="*80)
print("RESUMEN DE ALERTAS Y PROBLEMAS DETECTADOS")
print("="*80)

if len(alerts_df) > 0:
    print(f"Total de alertas: {len(alerts_df)}\n")
    print(alerts_df.to_string(index=False))
else:
    print("✓ No se detectaron alertas - El dataset está en buenas condiciones")
````

## 12. Resumen del Pipeline ML

````python
# Obtener resumen del pipeline
pipeline_summary = explorer.get_pipeline_summary()

print("\n" + "="*80)
print("RESUMEN DEL PIPELINE ML")
print("="*80)

print("\nCOLUMNAS NUMÉRICAS:")
print(f"  Cantidad: {len(pipeline_summary['numeric_cols'])}")
print(f"  Columnas: {pipeline_summary['numeric_cols']}")

print("\nCOLUMNAS CATEGÓRICAS:")
print(f"  Cantidad: {len(pipeline_summary['categorical_cols'])}")
print(f"  Columnas: {pipeline_summary['categorical_cols']}")

print("\nCOLUMNAS BOOLEANAS:")
print(f"  Cantidad: {len(pipeline_summary['boolean_cols'])}")
print(f"  Columnas: {pipeline_summary['boolean_cols']}")

print("\nCOLUMNAS TEMPORALES:")
print(f"  Cantidad: {len(pipeline_summary['datetime_cols'])}")
print(f"  Columnas: {pipeline_summary['datetime_cols']}")

print("\nCOLUMNAS DE TEXTO LIBRE:")
print(f"  Cantidad: {len(pipeline_summary['freetext_cols'])}")
print(f"  Columnas: {pipeline_summary['freetext_cols']}")

print("\nCOLUMNAS CON VALORES FALTANTES:")
print(f"  Cantidad: {len(pipeline_summary['cols_with_nulls'])}")
print(f"  Columnas: {pipeline_summary['cols_with_nulls']}")
````

## 13. Perfilado Completo del Dataset (Full EDA)

````python
# Ejecutar el EDA completo (todas las funciones en una sola llamada)
print("\n" + "="*80)
print("EJECUTANDO ANÁLISIS EXPLORATORIO COMPLETO")
print("="*80)

full_eda_results = explorer.run_full_eda()

print("\n" + "="*80)
print("RESULTADOS DISPONIBLES EN EL DICCIONARIO 'full_eda_results':")
print("="*80)
print(f"Claves disponibles ({len(full_eda_results)}):")
for i, key in enumerate(full_eda_results.keys(), 1):
    result = full_eda_results[key]
    result_type = type(result).__name__
    if isinstance(result, (pd.DataFrame, list, dict)):
        if isinstance(result, pd.DataFrame):
            print(f"  {i}. '{key}': DataFrame ({result.shape[0]}×{result.shape[1]})")
        elif isinstance(result, list):
            print(f"  {i}. '{key}': Lista ({len(result)} elementos)")
        elif isinstance(result, dict):
            print(f"  {i}. '{key}': Diccionario ({len(result)} claves)")
    else:
        print(f"  {i}. '{key}': {result_type}")

print("\n✓ ANÁLISIS EXPLORATORIO COMPLETO FINALIZADO")
````

````python
# Resumen final del análisis
print("\n" + "="*80)
print("RESUMEN FINAL DEL ANÁLISIS EXPLORATORIO")
print("="*80)

print("\n📊 INFORMACIÓN GENERAL DEL DATASET:")
print(f"  • Dimensiones: {df.shape[0]} filas × {df.shape[1]} columnas")
print(f"  • Tamaño en memoria: {df.memory_usage(deep=True).sum() / (1024**2):.2f} MB")
print(f"  • Tipos de datos únicos: {df.dtypes.nunique()}")

print("\n📈 DISTRIBUCIÓN DE COLUMNAS:")
print(f"  • Numéricas: {len(pipeline_summary['numeric_cols'])}")
print(f"  • Categóricas: {len(pipeline_summary['categorical_cols'])}")
print(f"  • Temporales: {len(pipeline_summary['datetime_cols'])}")
print(f"  • Booleanas: {len(pipeline_summary['boolean_cols'])}")

print("\n⚠️  PROBLEMAS DETECTADOS:")
print(f"  • Valores faltantes: {df.isna().sum().sum()} ({(df.isna().sum().sum() / (df.shape[0] * df.shape[1]) * 100):.2f}%)")
print(f"  • Filas duplicadas: {df.duplicated().sum()}")
print(f"  • Columnas con varianza baja: {len(low_variance_cols)}")
print(f"  • Valores atípicos detectados: {len(outliers_df)}")
print(f"  • Alertas totales: {len(alerts_df)}")

print("\n✅ ANÁLISIS COMPLETADO EXITOSAMENTE")
print("\nPróximos pasos recomendados:")
print("  1. Revisar valores faltantes y decidir estrategia de imputación")
print("  2. Analizar outliers y determinar si eliminar o transformar")
print("  3. Considerar transformaciones para variables no normales")
print("  4. Preparar los datos para modelado con data_cleaner.py")
````
