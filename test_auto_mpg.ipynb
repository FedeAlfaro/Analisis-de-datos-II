# Prueba de Funciones: DataCleaner y DataRegresion

**Dataset:** auto-mpg.csv

Este notebook valida todas las funciones disponibles en `data_cleaner.py` y `data_reggresion.py` usando el dataset Auto-MPG de 398 vehículos.

## 0. Importar Librerías

```python
import pandas as pd
import numpy as np
from pathlib import Path
import sys
import warnings
warnings.filterwarnings('ignore')

print("=" * 80)
print("PRUEBA DE FUNCIONES: data_cleaner y data_reggresion")
print("Dataset: auto-mpg.csv")
print("=" * 80)
```

## 1. Cargar Datos

```python
print("\n[1] CARGANDO DATOS...")
try:
    df = pd.read_csv("auto-mpg.csv")
    print(f"✓ Datos cargados exitosamente")
    print(f"  Dimensiones: {df.shape[0]} filas × {df.shape[1]} columnas")
    print(f"  Columnas: {list(df.columns)}")
    print(f"\nPrimeras 5 filas:")
    print(df.head())
except Exception as e:
    print(f"✗ Error al cargar datos: {e}")
```

## 2. Importar Módulos

```python
print("\n[2] IMPORTANDO MÓDULOS...")
try:
    from data_cleaner import DataCleaner
    from data_reggresion import DataRegresion
    print("✓ data_cleaner.py importado (DataCleaner)")
    print("✓ data_reggresion.py importado (DataRegresion)")
    
except Exception as e:
    print(f"✗ Error al importar módulos: {e}")
    import traceback
    traceback.print_exc()
```

## 3. Probar DataCleaner

```python
print("\n[3] PROBANDO DataCleaner...")
print("-" * 80)

try:
    # Crear instancia de DataCleaner
    df_copy = df.copy()
    
    # Primero crear instancia básica
    print("  ► Inicializando DataCleaner...")
    cleaner = DataCleaner(df_copy, verbose=True)
    print("    ✓ DataCleaner inicializado correctamente")
    
    # Mostrar métodos disponibles
    public_methods = [m for m in dir(cleaner) if not m.startswith('_') and callable(getattr(cleaner, m))]
    print(f"\n  Métodos disponibles en DataCleaner:")
    for method_name in public_methods:
        print(f"    - {method_name}()")
    
except Exception as e:
    print(f"✗ Error probando DataCleaner: {e}")
    import traceback
    traceback.print_exc()
```

### 3.1 Ejecutar Análisis Completo con DataCleaner

```python
print("\n  ► Ejecutando: run_full_analysis()")
try:
    results = cleaner.run_full_analysis()
    print(f"    ✓ Retorna Dict con keys: {list(results.keys())}")
    print(f"\n  Componentes del análisis:")
    for key in results.keys():
        value = results[key]
        if isinstance(value, pd.DataFrame):
            print(f"    • {key}: DataFrame {value.shape}")
        elif isinstance(value, dict):
            print(f"    • {key}: Dict con {len(value)} items")
        elif isinstance(value, list):
            print(f"    • {key}: List con {len(value)} items")
        else:
            print(f"    • {key}: {type(value).__name__}")
except Exception as e:
    print(f"    ⚠ Error: {str(e)[:100]}")
```

### 3.2 Explorar Resultados de DataCleaner

```python
# Ver análisis de correlaciones
if 'correlations' in results:
    print("\n[Correlations Analysis]:")
    print(results['correlations'].head())

# Ver análisis de cardinalidad
if 'cardinality' in results:
    print("\n[Cardinality Analysis]:")
    print(results['cardinality'][['column', 'n_unique', 'cardinality_level']].head())
```

## 4. Probar DataRegresion

```python
print("\n[4] PROBANDO DataRegresion...")
print("-" * 80)

try:
    # Crear instancia de DataRegresion
    # En auto-mpg.csv, 'mpg' es la columna objetivo (lo que queremos predecir)
    df_copy = df.copy()
    
    print("  ► Inicializando DataRegresion con target_col='mpg'...")
    regression = DataRegresion(df_copy, target_col='mpg', verbose=True)
    print("    ✓ DataRegresion inicializado correctamente")
    
    # Mostrar métodos disponibles
    public_methods = [m for m in dir(regression) if not m.startswith('_') and callable(getattr(regression, m))]
    print(f"\n  Métodos disponibles en DataRegresion:")
    for method_name in public_methods:
        print(f"    - {method_name}()")
    
except Exception as e:
    print(f"✗ Error probando DataRegresion: {e}")
    import traceback
    traceback.print_exc()
```

### 4.1 Dividir Datos (Train/Test)

```python
print("\n  ► Ejecutando: split_data()")
try:
    regression.split_data(test_size=0.2, random_state=42)
    print(f"    ✓ Datos divididos correctamente")
    print(f"    • Tamaño train: {regression.X_train.shape[0] if hasattr(regression, 'X_train') else 'N/A'}")
    print(f"    • Tamaño test: {regression.X_test.shape[0] if hasattr(regression, 'X_test') else 'N/A'}")
except Exception as e:
    print(f"    ⚠ Error: {str(e)[:100]}")
```

### 4.2 Comparar Modelos

```python
print("\n  ► Ejecutando: compare_models()")
try:
    comparison = regression.compare_models()
    if isinstance(comparison, pd.DataFrame):
        print(f"    ✓ Comparación completada: {comparison.shape}")
        print("\n[Model Comparison]:")
        print(comparison)
    else:
        print(f"    ✓ Retorna: {type(comparison).__name__}")
except Exception as e:
    print(f"    ⚠ Error: {str(e)[:100]}")
```

### 4.3 Entrenar Mejor Modelo

```python
print("\n  ► Ejecutando: train_best_model()")
try:
    best_result = regression.train_best_model()
    if isinstance(best_result, dict):
        print(f"    ✓ Modelo entrenado correctamente")
        print(f"    • Keys: {list(best_result.keys())}")
    else:
        print(f"    ✓ Retorna: {type(best_result).__name__}")
except Exception as e:
    print(f"    ⚠ Error: {str(e)[:100]}")
```

### 4.4 Resumen Final

```python
print("\n  ► Ejecutando: final_summary()")
try:
    summary = regression.final_summary()
    if isinstance(summary, dict):
        print(f"    ✓ Resumen generado correctamente")
        print("\n[Final Summary]:")
        for key, value in summary.items():
            print(f"  • {key}: {value}")
    else:
        print(f"    ✓ Retorna: {type(summary).__name__}")
except Exception as e:
    print(f"    ⚠ Error: {str(e)[:100]}")
```

### 4.5 Exportar Modelo (Opcional)

```python
print("\n  ► Ejecutando: export_model_artifact()")
try:
    output_path = regression.export_model_artifact("modelo_auto_mpg.pkl")
    print(f"    ✓ Modelo exportado a: {output_path}")
except Exception as e:
    print(f"    ⚠ Error: {str(e)[:100]}")
```
